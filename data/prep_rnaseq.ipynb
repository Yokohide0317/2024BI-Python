```python
import pandas as pd
```

```python
counts = pd.read_csv("./original/GSE52778_raw_counts_GRCh38.p13_NCBI.tsv", sep="\t", index_col=0)
counts.index = counts.index.astype(str)
counts
```

```python
annot = pd.read_csv("./original/Human.GRCh38.p13.annot.tsv", sep="\t", index_col=0)
annot.index = annot.index.astype(str)
annot
```

```python
count_annot = counts.merge(annot[["Symbol"]], left_index=True, right_index=True)
count_annot = count_annot.set_index("Symbol")
count_annot
```

```python
acc_list = [ "GSM1275862", "GSM1275863", "GSM1275864", "GSM1275865", "GSM1275866", "GSM1275867", "GSM1275868", "GSM1275869", "GSM1275870", "GSM1275871", "GSM1275872", "GSM1275873", "GSM1275874", "GSM1275875", "GSM1275876", "GSM1275877"]
title_list = [
    "N61311_untreated", "N61311_Dex", "N61311_Alb", "N61311_Alb_Dex",
    "N052611_untreated", "N052611_Dex", "N052611_Alb", "N052611_Alb_Dex",
    "N080611_untreated", "N080611_Dex", "N080611_Alb", "N080611_Alb_Dex",
    "N061011_untreated", "N061011_Dex", "N061011_Alb", "N061011_Alb_Dex"
]

sample_df = pd.DataFrame(index=acc_list, data={"title": title_list})
sample_df["treatment"] = sample_df["title"].apply(lambda x: "_".join(x.split("_")[1:]))
sample_df["cell_line"] = sample_df["title"].apply(lambda x: x.split("_")[0])

sample_df = sample_df[sample_df["treatment"].isin(["untreated", "Dex"])]
sample_df = sample_df.sort_values("treatment")
sample_df
```

```python
sample_df.reset_index(drop=True).to_csv("./csv/rnaseq_GSE52778_metadata.csv", sep=",", index=False)
```

```python
count_annot_renamed = count_annot[sample_df.index]
count_annot_renamed = count_annot_renamed.rename(columns=sample_df["title"])
count_annot_renamed.to_csv("./csv/rnaseq_GSE52778_counts.csv", sep=",", index_label="Gene")
count_annot_renamed
```
